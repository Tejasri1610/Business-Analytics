```python
import pandas as pd

df = pd.read_csv("Data_set 2 - Copy.csv")

df.head()
print("\n TASK-1 ")
print("\nðŸ”¹ Dataset Info:")
print(df.info())

print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

print("\nðŸ”¹ Summary Statistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("\n TASK-2 ")
gender_counts = df['gender'].value_counts().reset_index()
gender_counts.columns = ['gender', 'count']

plt.figure(figsize=(6, 4))
sns.barplot(data=gender_counts, x='gender', y='count', hue='gender', palette='pastel', legend=False)
plt.title('Gender Distribution - Bar Chart')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

gender_counts = df['gender'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel')
)
plt.title('Gender Distribution - Pie Chart')
plt.axis('equal')
plt.show()

```

```python
print("\n TASK-3 ")
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_cols:
    print(f"\n Statistics for {col}:")
    print(f" Mean: {df[col].mean():.2f}")
    print(f" Median: {df[col].median():.2f}")
    print(f" Standard Deviation: {df[col].std():.2f}")
```

```python
print("\n TASK-4 ")
preferred = df['Avenue'].value_counts()

print("\nðŸ“Œ Most Preferred Investment Avenues:")
print(preferred)

plt.figure(figsize=(8, 4))
sns.barplot(x=preferred.index, y=preferred.values, palette='pastel')
plt.title('Most Preferred Investment Avenues')
plt.xlabel('Investment Avenue')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
print("\n TASK-5 ")
reason_cols = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']
print(df[reason_cols].head(10))
```

```python
for col in reason_cols:
    print(f"\nReasons for choosing {col.replace('Reason_', '')}:")
    print(df[col].value_counts(dropna=True))
```

```python
print('Participants cited various reasons for their investment preferences. For Mutual Funds, common motivations included "diversification" and "professional management." Those who preferred Equity often mentioned "high returns" and "risk appetite." Fixed Deposits were favored for "safety" and "stable income," while Bonds appealed due to "government backing" and "medium-term planning."')
```

```python
print("\n TASK-6 ")
savings_col = 'What are your savings objectives?'

print("\n Sample Savings Objectives:")
print(df[savings_col].value_counts(dropna=True).head(10))
```

```python
all_text = " ".join(df[savings_col].dropna().astype(str).str.lower())

from collections import Counter
import re

words = re.findall(r'\b\w+\b', all_text)
word_counts = Counter(words)

print(" Top keywords in savings objectives:")
print(word_counts.most_common(10))
```

```python
plt.figure(figsize=(8, 4))
df[savings_col].value_counts().plot(kind='bar', color='skyblue')
plt.title('Savings Objectives of Participants')
plt.xlabel('Objective')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("Participants reported diverse savings objectives. The most frequently mentioned goals included retirement planning, children's education, buying a home, and emergency funds. This reflects a focus on both long-term financial security and short-term preparedness.")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import re
print("\n TASK-7 ")
info_col = 'Source'

print(df[info_col].value_counts(dropna=True).head(10))

all_info = " ".join(df[info_col].dropna().astype(str).str.lower())
info_words = re.findall(r'\b\w+\b', all_info)
info_counts = Counter(info_words)
print(info_counts.most_common(10))

plt.figure(figsize=(8, 4))
df[info_col].value_counts().plot(kind='bar', color='plum')
plt.title('Sources of Investment Information')
plt.xlabel('Information Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import re
print(" TASK-8 ")
df = pd.read_csv("Data_set 2 - Copy.csv")

def extract_number(value):
    try:
        return float(re.search(r'\d+\.?\d*', str(value)).group())
    except:
        return None

df['Duration'] = df['Duration'].apply(extract_number)
average_duration = df['Duration'].mean()
print(f"\n Average Investment Duration: {average_duration:.2f} years")
```

```python
print(" TASK-9 ")
expectations = df['Expect'].value_counts()
print(expectations)

expectations.plot(kind='pie', autopct='%1.1f%%')
plt.title("Investment Expectations")
plt.ylabel('')
plt.show()
```

```python
import seaborn as sns
print(" TASK-10 ")
df['Duration_Num'] = pd.to_numeric(df['Duration'], errors='coerce')
corr_columns = ['age', 'Duration_Num', 'Mutual_Funds', 'Equity_Market', 'Fixed_Deposits']
corr_columns = [col for col in corr_columns if col in df.columns]
correlation_matrix = df[corr_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
